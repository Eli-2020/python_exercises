```python
#Clean the Material "Wand" for all members in Harry Potter file
```

```python
harry = pd.read_csv("./Characters.csv", sep=';')
harry
```

```python
harry["Wand"] = harry["Wand"].astype(str)
```

```python
harry["Wand"] = harry["Wand"].apply(str.lower)
```

```python
harry["Wand"] = harry["Wand"].apply(lambda x: x.replace(",", ""))
```

```python
harry["Wand"] = harry["Wand"].apply(lambda x: x.split(" "))
```

```python
def give_me_wood_type(wand_list):
    if len(wand_list) >= 2:
        return wand_list[2]
    return None
```

```python
harry["Wand"] = harry["Wand"].apply(give_me_wood_type)
```
